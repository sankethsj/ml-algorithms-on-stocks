```python
import pandas as pd
from util import fetch_price, predict
```

```python
priceData = fetch_price()
```

```python
close_price = priceData['data']
len(close_price)
```

```python
window = 60
inputs = []
actual_outupt = []
for i in range(window, len(close_price)):
    inputs.append(close_price[i-window:i])
    actual_outupt.append(close_price[i])
```

```python
result = {
    'PrevClose': [],
    'Prediction': [],
    'Direction': [],
    'ActualClose': actual_outupt
}
for i in inputs:
    pred = predict(i)
    result['PrevClose'].append(pred['prevClose'])
    result['Prediction'].append(pred['pred'])
    result['Direction'].append(pred['direction'])
```

```python
res = pd.DataFrame(result)
```

```python
res
```

```python
def get_direction(row):
    return "Bullish" if row['ActualClose'] - row['PrevClose'] > 0 else "Bearish"
```

```python
res['ActualDirection'] = res.apply(get_direction, axis=1)
```

```python
res['ModelResult'] = res['Direction'] == res['ActualDirection']
```
